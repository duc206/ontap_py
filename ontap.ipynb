Khai báo một lớp đối tượng của lớp svk18A3 và gọi phương thức ghi đè để hiểu bản chất của đa hình 

```python
class SinhVien():
    def __init__(self, masv, hoten, lop):
        self.masv = masv
        self.hoten = hoten
        self.lop = lop 

    def inThongTinSinhVien(self):
        print("Mã sinh viên:", self.masv)
        print("Họ tên:", self.hoten)
        print("Lớp:", self.lop)

        
```
