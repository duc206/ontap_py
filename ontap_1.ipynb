Khai báo lớp SinhVien có các thuộc tính masv, hoten, lop

 - Viết phương thức info(), trả về thông tin của sinh viên

Định nghĩa lớp svk18A3 kế thừa từ lớp SinhVien

 - Ghi đè phương thức info()

Khai báo một lớp đối tượng của lớp svk18A3 và gọi phương thức ghi đè để hiểu bản chất của đa hình 

```python
class SinhVien():
    def __init__(self, masv, hoten, lop):
        self.masv = masv
        self.hoten = hoten
        self.lop = lop 

    def info(self):
        print("Mã sinh viên:", self.masv)
        print("Họ tên:", self.hoten)
        print("Lớp:", self.lop)

class svk18A3(SinhVien):
    def __init__(self, masv, hoten):
        super().__init__(masv, hoten, "K18A3")

    def info(self):
        return f"[SV K18A3] Mã SV: {self.masv}, Họ tên: {self.hoten}, Lớp: {self.lop}"
    

sv = svk18A3("24174600130", "Đỗ Minh Đức")
print(sv.info())

        
```
